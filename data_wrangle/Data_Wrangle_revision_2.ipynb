```python
def ncbi_enzyme_search(urls):
    enzymes = []
    
    for x in urls:
        
        response = requests.get(x)

        soup = BeautifulSoup(response.content, 'html.parser')
        text = soup.get_text()


        text = text.replace(".", "")
        text = text.replace(':','')
        text = text.replace(',','')
        text = text.replace('-',' ')
        text = text.replace("'",'')
        text = text.replace('(','')
        text = text.replace(')','')
        text = text.replace('=','')
        text = text.replace('"','')

        text = text.split()

        not_useful_ase = ['case', 'cases', 'increase', 'increases', 'increased', 'increasing','decrease', 'decreases', 'decreased','decreasing','release','releases','released','releasing','cease','ceases','ceased','ceasing','base','bases','based','disease','diseases','database','Cases','cerealbased','baseline','proteinbased','plantbased','increasesThe','grease','greases','greased','D3based', 'Disease', 'Diseases', 'Diseased', 'Increase','Increases','Increased','Decrease','Decreased','Decreases','Database','please','pleases','pleased''Please','Pleases','Pleased','diseases.','disease:','disease.','Disease:','Disease.']

        for x in text:
            ase = 'ase'
            if x not in not_useful_ase and ase in x:
                enzymes.append(x)

        
        
            
    return enzymes
```

```python
def kegg_api_find_hsa(gene_list):
    hsa_list_final = []
    for x in gene_list:
        URL = f'https://rest.kegg.jp/find/genes/{x}'
        r = requests.get(URL) 
        soup = bs4.BeautifulSoup(r.content, 
                             "lxml") 
        heading_object=soup.find_all( 'p' )
        
        soup_list = str(heading_object)

        soup_list = list(soup_list.split())
        
        hsa_list= []
        for y in soup_list:
            hsa_tag_pre = []
            hsa = []
            insert = 'hsa:'
            if insert in y:
                for z in y:
                    hsa_tag_pre.append(z)
                start = hsa_tag_pre.index('h')
                hsa = hsa_tag_pre[start:]
                hsa_tag = (''.join(hsa))
                hsa_list.append(hsa_tag)


            else:
                x=x
        if hsa_list != []:
            hsa_list_final.append(hsa_list[0])
            
    return hsa_list_final
```

```python
def gene_to_enzyme_class(hsa_list):
        classification_list=[]
        for hsa in hsa_list:
            URL = f'https://rest.kegg.jp/link/enzyme/{hsa}'
            r = requests.get(URL) 
            soup = bs4.BeautifulSoup(r.content, "lxml") 
            heading_object=soup.find_all( 'p' )

            if heading_object == []:
                print ('None')
            else:
                listed = str(heading_object)
                soup_list = list(listed.split())

                for x in soup_list:
                    if 'ec' in x:
                        ec_list =[]
                        for y in x:
                            ec_list.append(y)
                        ind = ec_list.index(':')
                classification = ec_list[ind+1]

                classification_list.append(classification)

        
        return classification_list
```

lets examine the questionable genes first, and then see if we can elminiate some genes for next query

1.) the entries with periods are said to be 'undergoing sequencing' at the NIH gene query, so lets drop those as they are not valid

2.) the entries starting with LINC are non protein coding sections, so they should dropped as well

3.) entries with dashes get hits when the the dashed part is removed

```python
def gene_cleaner(gene_list):
    clean_list = []
    for gene in gene_list:
        if '.' in gene:
            continue
        if 'LINC' in gene:
            continue
        if '-' in gene:
            gene_split = []
            for x in gene:
                gene_split.append(x)
            end = gene_split.index('-')
            clean = gene_split[:end]
            clean_list.append(''.join(clean))
        else:
            clean_list.append(gene)
        
            
    return clean_list
            

                
            
```
